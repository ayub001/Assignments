```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("df3.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.plot.scatter(x='a',y='b',c='red',s=50,figsize=(12,3))
```

```python
df['a'].plot.hist()
```

```python
plt.style.use('ggplot')
df['a'].plot.hist(alpha=0.5,bins=25)
```

```python
df[['a','b']].plot.box()
```

```python
df['d'].plot.kde()
```

```python
df['d'].plot.density(lw=5,ls='--')
```

```python
df[0:30].plot.area(alpha=0.4)
```

```python
f = plt.figure()
df[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
