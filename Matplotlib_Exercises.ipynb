```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
axis.plot(x,y)
axis.set_xlabel('x-axis')
axis.set_ylabel('y-axis')
axis.set_title('title')
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0,0,1,1])
axis2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
axis1.plot(x,y,color='black')
axis2.plot(x,y,color='red')
fig
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0,0,1,1])
axis2 = fig.add_axes([0.2,0.5,0.4,0.4])

#Large
axis1.set_xlabel('x')
axis1.set_ylabel('z')

axis1.plot(x,z)

#Inserted
axis2.set_xlabel('x')
axis2.set_ylabel('y')
axis2.set_title('zoom')

axis2.plot(x,y)
axis2.set_xlim(left=20,right=22)
axis2.set_ylim(bottom=30,top=50)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,z,color='r',lw=4)
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))
axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,z,color='r',lw=4)
```
